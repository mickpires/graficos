```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Filtro azul

azul = np.array([[1.5, 1.4, 1.3, 1.2, 1.1, 1. , 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1,0]]).T
azul_resultado = np.array([0,0,0,0.2,0.2,.5,1,2.2,4.4,8,14.5,28,37,51,65.8,84.8])

# filtro verde

verde = np.array([np.arange(1.3,0,-.09)])
verde  = verde.T
verde_resultado = np.array([0,0,0,0.2,.2,.4,.5,1,3,7.8,17,31,49,62.5,71])

# filtro vermelho

vermelho = np.array([[0.8 , 0.75, 0.7 , 0.65, 0.6 , 0.55, 0.5 , 0.45, 0.4, 0.35, 0.3,
       0.25, 0.2 , 0.15, 0.1 , 0.05,0]]).T
vermelho_resultado = np.array([0,0,0,0,.2,.2,.2,.4,.4,.4,.8,1.5,3.2,6,8.5,11,14])

# filtro amarelo
amarelo = np.array([np.arange(1.4,0.13,-.09)])
amarelo = amarelo.T
amarelo_resultado = np.array([0,0,.1,.3,.5,1,2,4,8,17.3,32,46.5,66.7,93.5,131])

# sem filtro

sem_filtro = np.array([np.arange(1.5,0.1,-.1)]).T
sem_filtro_resultado = np.array([0,.2,.4,.7,1,2,4,7.5,14.5,30,38.5,61,85,118])
```

```python
model_azul = LinearRegression()
model_azul.fit(azul,azul_resultado)

model_verde = LinearRegression()
model_verde.fit(verde,verde_resultado)

model_vermelho = LinearRegression()
model_vermelho.fit(vermelho,vermelho_resultado)

model_amarelo = LinearRegression()
model_amarelo.fit(amarelo,amarelo_resultado)

model_sem_filtro = LinearRegression()
model_sem_filtro.fit(sem_filtro,sem_filtro_resultado)
```

```python
# ax + b
print(f'a = {model_azul.coef_}')
print(f'b = {model_azul.intercept_}')
```
